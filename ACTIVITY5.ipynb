```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


file_path ='datasets\\Internet-Demographic.csv'
df = pd.read_csv(file_path)
df.columns = df.columns.str.strip()
df.info()
df.head()
```

 <h1> Task 1: Distribution of Birth Rates</h1>

```python
plt.figure(figsize=(10, 6))
sns.histplot(df["Birth rate"], bins=20, kde=True)
plt.title("Distribution of Birth Rates")
plt.xlabel("Birth Rate")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

 <h1>Task 2: Top 5 Countries with the Highest Internet Usage</h1>

```python
top_5_internet = df.nlargest(5, "Internet users")
print("Top 5 Countries with the Highest Internet Usage:")
print(top_5_internet[["Country Name", "Internet users"]])
```

 <h1>Task 3: Count of Countries by Income Group</h1>

```python
income_counts = df["Income Group"].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=income_counts.index, y=income_counts.values, palette="viridis")
plt.title("Count of Countries by Income Group")
plt.xlabel("Income Group")
plt.ylabel("Number of Countries")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

 <h1>Task 4: Box Plot of Internet Usage (Optional)</h1>

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y=df["Internet users"], color="skyblue")
plt.title("Box Plot of Internet Usage")
plt.ylabel("Internet Users (%)")
plt.grid(True)
plt.show()
```

 <h1>Task 5: Countries with No Internet Access</h1>

```python
no_internet = df[df["Internet users"] == 0]
print("Countries with No Internet Access:")
print(no_internet["Country Name"].tolist())
```

 <h1>Task 6: Average Birth Rate</h1>

```python
average_birth_rate = df["Birth rate"].mean()
print(f"Average Birth Rate: {average_birth_rate:.2f}")
```

 <h1>Task 7: Scatter Plot - Birth Rate vs. Internet Usage</h1>

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df["Birth rate"], y=df["Internet users"], alpha=0.7)
plt.title("Scatter Plot: Birth Rate vs. Internet Usage")
plt.xlabel("Birth Rate")
plt.ylabel("Internet Users (%)")
plt.grid(True)
plt.show()
```

 <h1>Task 8: Countries with Birth Rates Above 40</h1>

```python
high_birth_rate = df[df["Birth rate"] > 40]
print("Countries with Birth Rates Above 40:")
print(high_birth_rate[["Country Name", "Birth rate"]])
```

 <h1>Task 9: Pie Chart of Income Groups</h1>

```python
plt.figure(figsize=(8, 8))
income_counts.plot.pie(autopct='%1.1f%%', cmap="viridis", startangle=140)
plt.title("Proportion of Countries by Income Group")
plt.ylabel("")
plt.show()
```

 <h1>Task 10: Correlation Between Birth Rate and Internet Users</h1>

```python

correlation = df["Birth rate"].corr(df["Internet users"])
print(f"Correlation between Birth Rate and Internet Users: {correlation:.2f}")
if correlation > 0:
    print("Positive correlation")
elif correlation < 0:
    print("Negative correlation")
else:
    print("No correlation")
```
