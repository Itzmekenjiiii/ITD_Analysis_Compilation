```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Load the dataset
dengue_data = pd.read_csv('datasets\\dengue.csv')


dengue_data.head()
```

## 1. What is the trend of dengue cases over the years?

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=dengue_data, x="Year", y="Dengue_Cases", estimator="sum", ci=None, marker='o')
plt.title("Trend of Dengue Cases Over the Years")
plt.ylabel("Total Dengue Cases")
plt.xlabel("Year")
plt.grid(True)
plt.show()
```

## 2. What is the trend of dengue deaths over the years?

```python

plt.figure(figsize=(10, 6))
sns.lineplot(data=dengue_data, x="Year", y="Dengue_Deaths", estimator="sum", ci=None, marker='o', color='red')
plt.title("Trend of Dengue Deaths Over the Years")
plt.ylabel("Total Dengue Deaths")
plt.xlabel("Year")
plt.grid(True)
plt.show()
```

## 3. Which region has the highest number of dengue cases?

```python

plt.figure(figsize=(12, 6))
cases_by_region = dengue_data.groupby('Region')['Dengue_Cases'].sum().sort_values(ascending=False)
sns.barplot(x=cases_by_region.index, y=cases_by_region.values, palette="viridis")
plt.title("Total Dengue Cases by Region")
plt.ylabel("Total Dengue Cases")
plt.xlabel("Region")
plt.xticks(rotation=45)
plt.show()
```

## 4. Which region has the highest number of dengue deaths?

```python

plt.figure(figsize=(10, 6))
sns.lineplot(data=dengue_data, x="Year", y="Dengue_Deaths", estimator="sum", ci=None, marker='o', color='red')
plt.title("Trend of Dengue Deaths Over the Years")
plt.ylabel("Total Dengue Deaths")
plt.xlabel("Year")
plt.grid(True)
plt.show()
```

## 5. Is there a seasonal pattern in dengue cases?

```python

plt.figure(figsize=(12, 6))
cases_by_region = dengue_data.groupby('Region')['Dengue_Cases'].sum().sort_values(ascending=False)
sns.barplot(x=cases_by_region.index, y=cases_by_region.values, palette="viridis")
plt.title("Total Dengue Cases by Region")
plt.ylabel("Total Dengue Cases")
plt.xlabel("Region")
plt.xticks(rotation=45)
plt.show()
```

## 6. Is there a correlation between dengue cases and dengue deaths?

```python

plt.figure(figsize=(12, 6))
sns.boxplot(x='Month', y='Dengue_Cases', data=dengue_data, palette='Blues')
plt.title("Seasonal Pattern of Dengue Cases by Month")
plt.ylabel("Dengue Cases")
plt.xlabel("Month")
plt.xticks(rotation=45)
plt.show()
```

## 7. Which month has the highest average number of dengue cases?

```python

correlation, _ = pearsonr(dengue_data['Dengue_Cases'], dengue_data['Dengue_Deaths'])
print(f"Correlation between Dengue Cases and Dengue Deaths: {correlation:.2f}")

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Dengue_Cases', y='Dengue_Deaths', data=dengue_data)
plt.title("Correlation between Dengue Cases and Deaths")
plt.xlabel("Dengue Cases")
plt.ylabel("Dengue Deaths")
plt.grid(True)
plt.show()
```

## 8. Which year experienced the highest case fatality rate?

```python

dengue_data['CFR'] = (dengue_data['Dengue_Deaths'] / dengue_data['Dengue_Cases']) * 100
yearly_cfr = dengue_data.groupby('Year')['CFR'].mean()
max_cfr_year = yearly_cfr.idxmax()
print(f"The year with the highest Case Fatality Rate (CFR) is {max_cfr_year} with a CFR of {yearly_cfr[max_cfr_year]:.2f}%")

plt.figure(figsize=(10, 6))
sns.lineplot(data=yearly_cfr, marker='o', color='purple')
plt.title("Yearly Case Fatality Rate (CFR)")
plt.ylabel("CFR (%)")
plt.xlabel("Year")
plt.grid(True)
plt.show()
```

## 9. Are there any regions with consistently low or high case fatality rates?

```python

plt.figure(figsize=(10, 6))
sns.lineplot(data=dengue_data, x="Year", y="Dengue_Cases", estimator="sum", ci=None, marker='o')
plt.title("Trend of Dengue Cases Over the Years")
plt.ylabel("Total Dengue Cases")
plt.xlabel("Year")
plt.grid(True)
plt.show()
```

## 10. Are there significant spikes in dengue cases in certain regions?

```python

plt.figure(figsize=(12, 6))
cases_by_region = dengue_data.groupby('Region')['Dengue_Cases'].sum().sort_values(ascending=False)
sns.barplot(x=cases_by_region.index, y=cases_by_region.values, palette="viridis")
plt.title("Total Dengue Cases by Region")
plt.ylabel("Total Dengue Cases")
plt.xlabel("Region")
plt.xticks(rotation=45)
plt.show()
```

## 11. What is the distribution of dengue cases per region?

```python

plt.figure(figsize=(12, 6))
deaths_by_region = dengue_data.groupby('Region')['Dengue_Deaths'].sum().sort_values(ascending=False)
sns.barplot(x=deaths_by_region.index, y=deaths_by_region.values, palette="Reds")
plt.title("Total Dengue Deaths by Region")
plt.ylabel("Total Dengue Deaths")
plt.xlabel("Region")
plt.xticks(rotation=45)
plt.show()
```

## 12. What is the distribution of dengue deaths per region?

```python

correlation, _ = pearsonr(dengue_data['Dengue_Cases'], dengue_data['Dengue_Deaths'])
print(f"Correlation between Dengue Cases and Dengue Deaths: {correlation:.2f}")

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Dengue_Cases', y='Dengue_Deaths', data=dengue_data)
plt.title("Correlation between Dengue Cases and Deaths")
plt.xlabel("Dengue Cases")
plt.ylabel("Dengue Deaths")
plt.grid(True)
plt.show()
```

## 13. Are there any regions where dengue deaths increased despite fewer cases?

```python

monthly_avg_cases = dengue_data.groupby('Month')['Dengue_Cases'].mean().sort_values(ascending=False)
print("Average Dengue Cases per Month:")
print(monthly_avg_cases)

plt.figure(figsize=(12, 6))
sns.barplot(x=monthly_avg_cases.index, y=monthly_avg_cases.values, palette='viridis')
plt.title("Average Dengue Cases per Month")
plt.ylabel("Average Dengue Cases")
plt.xlabel("Month")
plt.xticks(rotation=45)
plt.show()
```

## 14. What is the overall trend of dengue cases during the rainy season versus dry season?

```python

region_cfr = dengue_data.groupby('Region')['CFR'].mean().sort_values()
print("Average Case Fatality Rate by Region:")
print(region_cfr)

plt.figure(figsize=(12, 6))
sns.barplot(x=region_cfr.index, y=region_cfr.values, palette='coolwarm')
plt.title("Average Case Fatality Rate (CFR) by Region")
plt.ylabel("CFR (%)")
plt.xlabel("Region")
plt.xticks(rotation=45)
plt.show()
```

## 15. Is there a significant correlation between year and case fatality rate?

```python

plt.figure(figsize=(14, 8))
sns.lineplot(data=dengue_data, x='Year', y='Dengue_Cases', hue='Region', estimator='sum')
plt.title("Dengue Cases Trend by Region Over the Years")
plt.ylabel("Dengue Cases")
plt.xlabel("Year")
plt.legend(title="Region", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

## 16. What is the percentage change in dengue cases year-over-year?

```python

plt.figure(figsize=(12, 6))
sns.boxplot(x='Region', y='Dengue_Deaths', data=dengue_data)
plt.title("Distribution of Dengue Deaths by Region")
plt.ylabel("Dengue Deaths")
plt.xlabel("Region")
plt.xticks(rotation=45)
plt.show()
```

## 17. How effective have interventions been in reducing dengue cases?

```python

dengue_data['Deaths_Per_Case'] = dengue_data['Dengue_Deaths'] / dengue_data['Dengue_Cases']
increased_deaths = dengue_data.sort_values(by='Deaths_Per_Case', ascending=False)
increased_deaths[['Year', 'Region', 'Dengue_Cases', 'Dengue_Deaths', 'Deaths_Per_Case']].head(10)
```

## 18. Correlation between Year and CFR

```python

correlation_year_cfr, _ = pearsonr(dengue_data['Year'], dengue_data['CFR'])
print(f"Correlation between Year and CFR: {correlation_year_cfr:.2f}")
```

## 19. Year-over-Year Percentage Change in Dengue Cases

```python

yearly_cases = dengue_data.groupby('Year')['Dengue_Cases'].sum()
yearly_pct_change = yearly_cases.pct_change() * 100
print("Year-over-Year Percentage Change in Dengue Cases:")
print(yearly_pct_change)
```

## 20. Yearly Dengue Cases and Impact of Interventions

```python

plt.figure(figsize=(10, 6))
sns.lineplot(data=yearly_cases, marker='o', color='green')
plt.title("Yearly Dengue Cases and Impact of Interventions")
plt.ylabel("Total Dengue Cases")
plt.xlabel("Year")
plt.grid(True)
plt.show()
```
